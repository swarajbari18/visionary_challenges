```python
import pandas as pd
import numpy as np
import datetime


def passage_type(df):
    return df[0][0]

def row_index(df):
    index_req = 0
    for element in df[0][:5]:
        if type(element)==str:
            if 'quest' in element.lower():
                return index_req
        index_req += 1
    return

def date_time_search(df, row_inde):
    '''
    Return the result in (row, col) format, colllect it as such tuple form in only one variable
    also put row in df.iloc[]
    and col in df[]'''
    for i in range(row_inde):
        col=0
        row = i
        for o in df.iloc[i]:
            if type(o)==datetime.datetime:
                return (row, col)
            
            col += 1
    return

def no_of_col(df):
    return len(df.columns)

def no_of_rows(df):
    return len(df)

def student_date_dictionary(df, rowcol, row_inde):
    '''
    rowcol is a tuple from date_time_searc() function
    row_inde is output of row_index() function'''
    date = []
    for i in df.iloc[rowcol[0]][rowcol[1]:]:
        date.append(i)
        
    students = []    
    for j in df.iloc[row_inde][rowcol[1]:]:
        students.append(j)
    
    diction = dict(zip(students, date))
    del date
    del students
    
    return diction

def clean_datframe(df):

    row_inde = row_index(df)
    
    rowcol = date_time_search(df, row_inde)
    total_col = no_of_col(df)
    dates = student_date_dictionary(df, rowcol, row_inde)
    df2 = df.iloc[row_inde+1:].copy()
    df2.columns = df.iloc[row_inde]

    return df2

def data_creation(df, rowcol):
    students = df.columns[rowcol[1]:]
    
    data = {'student_id':[i for i in students for j in range(len(df))],
           'student_answer':[j for i in range(rowcol[1], len(df.columns)) for j in df.iloc[:,i]],
           'correct_answer':[j for i in students for j in df['Answer']],
           'question_no':[j for i in students for j in df['Question']],
           'question_type':[j for i in students for j in df['Question Type']]}
    return data

def dataframing(data, sheet_name, passage_ttpe, test_type):
    df = pd.DataFrame(data)
    df['age'] = df['student_id'].apply(age_calculator)
    df['test_type'] = test_type
    df['passage_name'] = sheet_name
    df['passage_type'] = passage_ttpe 
    return df

def student_birthday():
    dd = pd.read_excel(r"D:\Machine_Learning\Dicey Tech\DiceyTech x Target Tuition (Official)\student profiles.xlsx")
    data_dict = dd.set_index('Student ID')['Birthday'].to_dict()
    #print(data_dict)
    del dd
    return data_dict

def age_calculator(x):
    '''
    dates student_dates_dictionary() function se
    data_dict () us function se
    df humara new generated df hai
    '''
    global data_dict
    global dates
    try:
        age = dates[x]-data_dict[x]
        age = age.days/365
        return age
    except:
        return np.NaN
    
    
    
    
    
def data_processor_proto2(df,sheets,test_type):
    
    sheets = sheets
    dataframes =[]
    for sheet in sheets:
        df1 = df[sheet].copy()
        row_inde = row_index(df1)
        passage = passage_type(df1)
        rowcol = date_time_search(df1, row_inde)
        global dates
        dates = student_date_dictionary(df1,rowcol, row_inde)
        df1 = clean_datframe(df1)
        data = data_creation(df1, rowcol)
        del df1
        df2 = dataframing(data, sheet, passage, test_type)
        dataframes.append(df2)
    
    return dataframes
```

```python
import spacy
import os
import sys
from collections import Counter
nlp = spacy.load('en_core_web_sm')
```

```python
#  WENT TO THE FOLDER WHERE I ALREADY HAD THE THE TEXT DAT, 
# SEPEREATED BETWEEN PASSAGE TEXT AND THE QUESTION WERE SEPERATED IN A JSON FILE I CREATED
folder = r'D:\Machine_Learning\Dicey Tech\Project\data\Workshet'
file_list = os.listdir(folder)
print(file_list)
```

```python
#  custom function i created to process the passage text data
import re
import string
punct = string.punctuation


def nlp_processor_for_passage_text(file_list, punct):

    dictionary_passage = dict()
    for file_name in file_list:
        file = open(f'D:\\Machine_Learning\\Dicey Tech\\Project\\data\\Workshet\\{file_name}', encoding='utf-8')
        data = file.read()
        file.close()
        data=data.replace('\n',' ').lower()
        data = re.sub(f'[{punct}]', '', data)
        passge_length = len(data)
        doc = nlp(data)

        no_of_words = len(doc)
        no_of_entities = len(doc.ents)

        no_of_sentence = 0
        no_of_words = 0
        len_of_sent = 0
        for sents in doc.sents:
            no_of_sentence += 1
            no_of_words += len(sents)
            len_of_sent += len(sents.text)
        #     print(len(sents))
        if no_of_sentence == 0:
            avg_length = 0
            avg_words = 0
        else:
            avg_words = no_of_words/no_of_sentence
            avg_length = len_of_sent/no_of_sentence




        lst = []
        for word in doc:
            tag = word.tag_
            lst.append(tag)
            lst.append(word.dep_)
        #     print(word,'----->', tag,'------->', spacy.explain(tag), '------->', word.dep_)

        features = Counter(lst)
        file_title = file_name.split('.')[0]
        dictionary_passage[f'{file_title}'] = features
        dictionary_passage[f'{file_title}']['passage_length'] = passge_length
        dictionary_passage[f'{file_title}']['no_of_words'] = no_of_words
        dictionary_passage[f'{file_title}']['no_of_sentances'] = no_of_sentence
        dictionary_passage[f'{file_title}']['avg_len_of_sentenc'] = avg_length
        dictionary_passage[f'{file_title}']['avg_no_of_words_per_sentence'] = avg_words
        dictionary_passage[f'{file_title}']['no_of_entities'] = no_of_entities

        del features

    return dictionary_passage
```

```python
dictionary_passage = nlp_processor_for_passage_text(file_list, punct)
dictionary_passage
# these are all the features extracted from th passage text
```

NOW TO PROCESS THIS DATA I CREATED A NEW CUSTOM FUNCTION FOR TIS AND RAN IT AND THEN STORED THE RESULLT IN TO A JSON FILE

```python
def nlp_processor_question_answer(data):
    
    data=data.replace('\n',' ').lower()
    data = re.sub(f'[{punct}]', '', data)
    passge_length = len(data)
    doc = nlp(data)
    no_of_words = len(doc)
    no_of_entities = len(doc.ents)
    lst = []
    for word in doc:
        tag = word.tag_
        lst.append(tag)
        lst.append(word.dep_)
    #     print(word,'----->', tag,'------->', spacy.explain(tag), '------->', word.dep_)

    features = Counter(lst)
    features['question_length'] = passge_length
    features['no_of_words'] = no_of_words
    features['no_of_entities'] = no_of_entities
    return features
```

```python

for test_type in py_di:
    
    for passage in py_di[test_type]:
        for question_no in py_di[test_type][passage]:
            for i in range(len(py_di[test_type][passage][question_no])):
                features = nlp_processor_question_answer(py_di[test_type][passage][question_no][i])
                py_di[test_type][passage][question_no][i] = features
                
```
